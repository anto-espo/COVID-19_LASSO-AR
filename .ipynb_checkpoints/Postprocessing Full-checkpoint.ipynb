```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import scipy as sp
import matplotlib.pyplot as plt
import re
from os import listdir
from os.path import isfile, join
from scipy import stats
from matplotlib.ticker import MaxNLocator

import warnings
warnings.filterwarnings('ignore')
```

```python
important = ['TLR3','TBPL2','GBP2','GBC4','N4BP2','AURKA','RRM2','TXK']
```

```python
df = pd.read_csv("./Results/rules_males_full.csv")
```

```python
#Split the string by reading the left hand and the right hand sides 
df['lhs'] = df['rules'].apply(lambda x : re.search(r"\{(.+)\} ", x).group(1))
df['rhs'] = df['rules'].apply(lambda x : re.search(r" \{(.+)\}", x).group(1))

#drop the rules column
df.drop(columns='rules',inplace = True)
#sort the columns differently 
df = df[['lhs','rhs','support','confidence','lift','count','coverage','pvalue_test']]

df.reset_index(inplace=True)
df.drop(columns='index', inplace=True)
df['lhs'] = df['lhs'].str.replace('=1','')
df['lhs'] = df['lhs'].str.replace('KRTAP27.1_1','KRTAP27-1_1')
df['count'] = df['count'].astype('int32')
df = df.sort_values(by=['support'], ascending=False)
```

```python
minconf_1 = 0.8
minsup_1 = 0.08

df[(df['confidence']>=minconf_1) 
                   & (df['support']>=minsup_1)
                   & (df['rhs']=='grading=1')
                  ]
```

```python
minconf_0 = 0.8
minsup_0 = 0.12

df[(df['confidence']>=minconf_0) 
                   & (df['support']>=minsup_0)
                   & (df['rhs']=='grading=0')
                  ]
```

# Importo geni di suscettibilità e protezione

```python
feature_importance = pd.read_csv('./Data/Male/relevant genes in males.csv', sep=';')
```

```python
feature_importance['complete'] = feature_importance['complete'].astype('str')
feature_importance.loc[feature_importance['Type']=='Common Homozygous']['complete'] = feature_importance[feature_importance['Type']=='Common Homozygous']['complete'].apply(lambda x :  x +'_homo')
feature_importance.loc[feature_importance['Type']=='Common Hemizygous']['complete'] = feature_importance[feature_importance['Type']=='Common Hemizygous']['complete'].apply(lambda x :  x +'_hemi')
feature_importance['Feature Importance'] = feature_importance['Feature Importance'].str.replace(',','.')
feature_importance['Feature Importance'] = pd.to_numeric(feature_importance['Feature Importance'],errors='coerce')
```

```python
feature_importance.complete = feature_importance.apply(lambda row: row['complete'] + '_hemi' if row['Type'] == 'Common Hemizygous' else row['complete'], axis=1)
```

```python
sus = feature_importance[feature_importance['Feature Importance']>=0]['complete'].tolist()
#sus = sus +['IRX5_1']
```

```python
prot = feature_importance[feature_importance['Feature Importance']<0]['complete']
```

```python
df80_1 = df[(df['confidence']>=minconf_1) 
                                          & (df['rhs']=='grading=1') 
                                          & (df['support']>=minsup_1)]

df80_1['mixed'] = df80_1.lhs.apply(lambda x : np.setdiff1d(x.split(','),sus))
```

```python
count= 0
for i in range(len(df80_1)):
    if df80_1.iloc[i]['mixed'].size > 0:
        print(df80_1.iloc[i]['mixed'])
        count+= 1
print(count)
print(len(df80_1))
```

```python
plt.figure(figsize=(12,12))

plot = plt.scatter(df80_1['lhs'], df80_1['count'], c=df80_1['confidence'], cmap='Reds')
plt.clf()
cbar = plt.colorbar(plot)
cbar.ax.set_title("confidence")

g=sns.barplot(x='count',y='lhs',data=df80_1, palette='Reds', hue='confidence', dodge=False)

g.set_title('Association Rules FEMALES minconf={minconf:.2f} minsup={minsup:.2f} grading=1'.format(minconf=minconf_1,minsup=minsup_1),pad=20)

locs, labels=plt.yticks()
for i in range(len(df80_1)):
    plt.annotate('{:1.2f}'.format(df80_1.iloc[i]['confidence']), xy=(df80_1.iloc[i]['count']/2,locs[i]), ha='center', va='center', fontsize=11)

    if(len(df80_1.iloc[i].mixed) >0):
        g.get_yticklabels()[i].set_fontweight("bold")
#plt.tick_params(labelsize=15)
plt.xticks(np.arange(0, max(df80_1['count'])+1, 5))
    
g.legend_.remove()
plt.tight_layout()
#plt.savefig('../../Plots/female_common_sus.png', dpi=150)
plt.show()
```

```python
df80_0 = df[(df['confidence']>=minconf_0) 
                                          & (df['rhs']=='grading=0') 
                                          & (df['support']>=minsup_0)]

df80_0['mixed'] = df80_0.lhs.apply(lambda x : np.setdiff1d(x.split(','),prot))
```

```python
count= 0
for i in range(len(df80_0)):
    if df80_0.iloc[i]['mixed'].size > 0:
        print(df80_0.iloc[i]['mixed'])
        count+= 1
print(count)
print(len(df80_0))
```

```python
plt.figure(figsize=(30,35))

plot = plt.scatter(df80_0['lhs'], df80_0['count'], c=df80_0['confidence'], cmap='Reds')
plt.clf()
cbar = plt.colorbar(plot)
cbar.ax.set_title("confidence")

g=sns.barplot(x='count',y='lhs',data=df80_0, palette='Reds', hue='confidence', dodge=False)

g.set_title('Association Rules FEMALES minconf={minconf:.2f} minsup={minsup:.2f} grading_adj_age=0'.format(minsup=minsup_0,minconf=minconf_0),pad=20)

locs, labels=plt.yticks()
for i in range(len(df80_0)):
    plt.annotate('{:1.2f}'.format(df80_0.iloc[i]['confidence']), xy=(df80_0.iloc[i]['count']/2,locs[i]), ha='center', va='center', fontsize=15)
    
    if(len(df80_0.iloc[i].mixed) >0):
        g.get_yticklabels()[i].set_fontweight("bold")
    
plt.xticks(np.arange(0, max(df80_0['count'])+1, 4))
plt.tick_params(labelsize=16)

g.legend_.remove()
plt.tight_layout()
#plt.savefig('../../Plots/Female_GC_Homo_0.pdf', dpi=150)
plt.show()
```

#### Protection

```python
pat0 = pat[pat['grading']==0]
```

```python
grading0 = df[((df['confidence']>=minconf_0) 
                                          & (df['rhs']=='grading=0') 
                                          & (df['support']>=minsup_0))]
grading0['lhs'] = grading0['lhs'].str.replace('.','-')

#grading0 = grading0[0:10]
#grading_adj_age0['lhs'] = grading_adj_age0['lhs'].str.replace('KRTAP5;10','KRTAP5-10')
```

# Count rules in common between genes

##### Grading = 0

```python
genes = set([])
genes2 = set([])

for rule in df80_0['lhs']:
    elements = rule.split(',')
    for e in elements:
        genes.add(e)
        genes2.add(e)
        
genes = list(genes)
genes2 = list(genes2)
        
matrix = np.full((len(genes),len(genes)),0)

for i in range(len(genes)):
    for j in range(len(genes)): 
        #inserisco nella casella della matrice il numero di regole che contengono entrambi i geni
        matrix[i,j] = len(df80_0[(df80_0['lhs'].str.contains(genes[i]))
                                                 & (df80_0['lhs'].str.contains(genes[j])) ])
```

```python
plt.figure(figsize=(12,10))
mask = np.triu(np.ones_like(matrix, dtype=bool), k=1)
sns.heatmap(matrix, xticklabels=genes, yticklabels=genes, annot=True, mask=mask, cmap='Blues')
```

##### Grading = 1

```python
genes = set([])
genes2 = set([])

for rule in df80_1['lhs']:
    elements = rule.split(',')
    for e in elements:
        genes.add(e)
        genes2.add(e)
        
genes = list(genes)
genes2 = list(genes2)
        
matrix = np.full((len(genes),len(genes)),0)

for i in range(len(genes)):
    for j in range(len(genes)): 
        #inserisco nella casella della matrice il numero di regole che contengono entrambi i geni
        matrix[i,j] = len(df80_1[(df80_1['lhs'].str.contains(genes[i]))
                                                 & (df80_1['lhs'].str.contains(genes[j])) ])
```

# Pazienti spiegati con X regole

```python
pat = pd.read_csv('./Data/AssociationRules/All_Male_bool.csv')
```

```python
pat0 = pat[pat['grading']==0]
```

```python
grading0 = df[((df['confidence']>=minconf_0) 
                                          & (df['rhs']=='grading=0') 
                                          & (df['support']>=minsup_0))]

grading0['mixed'] = grading0.lhs.apply(lambda x : np.setdiff1d(x.split(','),prot))
grading0['important'] = grading0.lhs.apply(lambda x : np.intersect1d(x.split(','),important))

grading0['lhs'] = grading0['lhs'].str.replace('.','-')
#grading_adj_age0['lhs'] = grading_adj_age0['lhs'].str.replace('KRTAP5;10','KRTAP5-10')
```

```python
series = []
genes = []
patients = []

for i in range(len(grading0)):
    genes.append(grading0.iloc[i]['lhs'].split(','))
    series.append(pd.Series((pat0[genes[i]].values == 1).all(axis=1), name='bools'))
    patients.append(pat0[series[i].values].index.values)
```

```python
distincts_per_rule = []

for i in range(1,len(patients)):
    distinct = patients[0]
    for j in range(1,i):
        distinct = np.union1d(distinct,patients[j])
    distincts_per_rule.append(len(distinct))
    
#print("Le regole più significative coprono: ",len(distinct), " pazienti non severi distinti su ",len(pat0))
#print("Corrispondente al {perc:.2f}% dei pazienti gravi".format(perc=(len(distinct)/len(pat0))*100))
#print("Corrispondente al {perc:.2f}% del totale".format(perc=(len(distinct)/len(pat))*100))
```

```python
sns.set_theme()

plt.figure(figsize=(20,10))
ax = sns.lineplot(x=range(1,len(patients)), y=distincts_per_rule)
ax.set_xlabel('Number of Rules', fontsize=18)
ax.set_ylabel('Distinct Patients', fontsize=18)
ax.tick_params(labelsize=15)
ax.set_xticks(np.arange(0, len(patients)+1, 4))
ax.set_yticks(np.arange(40, max(distincts_per_rule)+1, 4))
ax.set_xlim(0,40)
ax.set_title('Number of distinct male patients covered by protection rules', fontsize=20)
plt.plot(32,distincts_per_rule[31],'ok', markersize=10)
plt.tight_layout()
plt.savefig('./Plots/distinct_patients_male_protection.png', dpi=150)
```
