```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import math
import scipy as sp
```

# Phenotypes filtering

```python
pheno = pd.read_csv('./Data/phenotypes_1319.csv', sep=';')
```

```python
pheno
```

## Filtro per Etnia

```python
pheno['Ethnicity'] = pheno['Ethnicity (white=1, black=2, asian=3, hispanic=4)']
```

```python
pheno = pheno[(pheno['Ethnicity']=='1') | (pheno['Ethnicity'].isna())]
```

## Discard patients without grading

```python
pheno = pheno[pheno['grading_1319_adj_a_s'].notna()]
```

```python
pheno
```

### Remove patients without age

```python
pheno = pheno[pheno['Age'].notna()]
```

# Load patients datasets

Load common datasets

```python
gc_homo = pd.read_csv('./Data/data_gc_homo_poly_unique_1319.csv')
gc = pd.read_csv('./Data/data_gc_poly_unique_1319.csv')
gc_hemi = gc.copy()
```

Load rare datasets

```python
rare_AD = pd.read_csv('./Data/data_AD_rare_1319.csv')
#rare_AD.rename(columns={'Unnamed: 0_rare_AD':'Unnamed: 0'}, inplace=True)
rare_AR = pd.read_csv('./Data/data_AR_rare_1319.csv')
#rare_AR.rename(columns={'Unnamed: 0_rare_AR':'Unnamed: 0'}, inplace=True)
rare_hemi = rare_AD.copy()
```

Select Used Features

```python
selected = pd.read_csv('./Data/Male/relevant genes in males.csv', sep=';')
```

```python
selected = selected[['complete','Feature Importance','Type']]
selected.rename(columns={'complete':'feature'},inplace=True)
selected.dropna(inplace=True)
```

Select features for each dataset

```python
selected_gc = selected[selected['Type'] == 'Common Haplotypes']['feature'].tolist() + ['Unnamed: 0']
selected_gc_homo = selected[selected['Type'] == 'Common Homozygous']['feature'].tolist() + ['Unnamed: 0_homo']
selected_ad = selected[selected['Type'] == 'rare_AD']['feature'].tolist() + ['Unnamed: 0_rare_AD']
selected_ar = selected[selected['Type'] == 'rare_AR']['feature'].tolist() + ['Unnamed: 0_rare_AR']
selected_hemi_rare = selected[selected['Type'] == 'rare Hemizygous']['feature'].tolist() + ['Unnamed: 0_rare']
selected_gc_hemi = selected[selected['Type'] == 'Common Hemizygous']['feature'].tolist() + ['Unnamed: 0']
```

```python
gc = gc[selected_gc]

gc_homo.columns = gc_homo.columns.map(lambda x: str(x) + '_homo')
gc_homo = gc_homo[selected_gc_homo]

gc_hemi = gc_hemi[selected_gc_hemi]
gc_hemi.columns = gc_hemi.columns.map(lambda x: str(x) + '_hemi')

rare_AR.columns = rare_AR.columns.map(lambda x: str(x) + '_rare_AR')
rare_AR = rare_AR[selected_ar]

rare_AD.columns = rare_AD.columns.map(lambda x: str(x) + '_rare_AD')
rare_AD = rare_AD[selected_ad]

rare_hemi.columns = rare_hemi.columns.map(lambda x: str(x) + '_rare')
rare_hemi = rare_hemi[selected_hemi_rare]
```

Merge common datasets

```python
gc_homo.rename(columns={"Unnamed: 0_homo": "key"}, inplace=True)
gc_homo.set_index('key',inplace=True)

gc.rename(columns={"Unnamed: 0": "key"}, inplace=True)
gc.set_index('key',inplace=True)

gc_hemi.rename(columns={"Unnamed: 0_hemi": "key"}, inplace=True)
gc_hemi.set_index('key', inplace=True)

pheno.rename(columns={"UsedSampleCode":"key"}, inplace=True)
pheno.set_index('key',inplace=True)
```

```python
common = gc.join(gc_homo, on='key')
common = common.join(gc_hemi, on='key')
```

Merge rare datasets

```python
rare_AR.rename(columns={"Unnamed: 0_rare_AR": "key"}, inplace=True)
rare_AD.rename(columns={"Unnamed: 0_rare_AD": "key"}, inplace=True)
rare_hemi.rename(columns={"Unnamed: 0_rare": "key"}, inplace=True)
rare_AR.set_index('key',inplace=True)
rare_AD.set_index('key',inplace=True)
rare_hemi.set_index('key',inplace=True)
```

```python
rare = rare_AR.join(rare_AD, on='key')#, lsuffix='_rare_AR', rsuffix='_rare_AD')
rare = rare.join(rare_hemi, on='key')
```

Merge all datasets together

```python
all_features = rare.join(common, on='key')
```

```python
all_features
```

# Drop mismatching patients

```python
all_features = all_features[all_features.index != 'COV2925-1241_hg38']
all_features = all_features[all_features.index != 'COV2928-1242_hg38']
all_features = all_features[all_features.index != 'COV3204-1334_hg38']
all_features = all_features[all_features.index != 'COV3196-1326_hg38']
all_features = all_features[all_features.index != 'COV2735-1159_hg38']
all_features = all_features[all_features.index != 'COV3211-1341_hg38']
all_features = all_features[all_features.index != 'COV2233-939_hg38']

all_features.index = all_features.index.str.replace('_hg38','')
```

```python
grading = pheno[['grading_1319_adj_a_s','PatientID','Gender (M=0, F=1)']]

df = all_features.join(grading, on='key')
df.rename(columns={'grading_1319_adj_a_s':'grading'}, inplace=True)

df_controllo = df.copy()
df_controllo.dropna(inplace=True)

df = df[df['grading']!='none']
df.dropna(inplace=True)
df
```
