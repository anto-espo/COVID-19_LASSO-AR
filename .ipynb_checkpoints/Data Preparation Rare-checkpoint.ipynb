```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import math
import scipy as sp
```

# Phenotypes filtering

```python
pheno = pd.read_csv('./Data/phenotypes_1319.csv', sep=';')
```

```python
pheno
```

## Filtro per Etnia

```python
pheno['Ethnicity'] = pheno['Ethnicity (white=1, black=2, asian=3, hispanic=4)']
```

```python
pheno = pheno[(pheno['Ethnicity']=='1') | (pheno['Ethnicity'].isna())]
```

## Discard patients without grading

```python
pheno = pheno[pheno['grading_1319_adj_a_s'].notna()]
```

```python
pheno
```

### Remove patients without age

```python
pheno = pheno[pheno['Age'].notna()]
```

# Load rare datasets male patients

```python
rare_AR = pd.read_csv('./Data/data_AR_rare_1319.csv')
rare_AD = pd.read_csv('./Data/data_AD_rare_1319.csv')
```

Merge rare datasets

```python
rare_AR.rename(columns={"Unnamed: 0": "key"}, inplace=True)
rare_AD.rename(columns={"Unnamed: 0": "key"}, inplace=True)
rare_AR.set_index('key',inplace=True)
rare_AD.set_index('key',inplace=True)
pheno.rename(columns={"UsedSampleCode":"key"}, inplace=True)
pheno.set_index('key',inplace=True)
```

```python
rare = rare_AR.join(rare_AD, on='key', lsuffix='_rare_AR', rsuffix='_rare_AD')
```

```python
rare_hemi = rare_AD.copy()
rare_hemi.columns = rare_hemi.columns.map(lambda x: str(x) + '_rare')
rare_hemi
```

```python
rare = rare.join(rare_hemi)
```

# Drop mismatching patients

```python
rare = rare[rare.index != 'COV2925-1241_hg38']
rare = rare[rare.index != 'COV2928-1242_hg38']
rare = rare[rare.index != 'COV3204-1334_hg38']
rare = rare[rare.index != 'COV3196-1326_hg38']
rare = rare[rare.index != 'COV2735-1159_hg38']
rare = rare[rare.index != 'COV3211-1341_hg38']
rare = rare[rare.index != 'COV2233-939_hg38']

rare.index = rare.index.str.replace('_hg38','')
```

```python
grading = pheno[['grading_1319_adj_a_s','Gender (M=0, F=1)']]

df = rare.join(grading, on='key')
df.rename(columns={'grading_1319_adj_a_s':'grading'}, inplace=True)
df = df[df['grading']!='none']
df.dropna(inplace=True)
df
```

# Import selected features AD, AR and hemi

```python
selected_AR = pd.read_csv('./Data/Male/genes_AR_male.csv',sep=';')
selected_AD = pd.read_csv('./Data/Male/genes_AD_male.csv',sep=';')
selected_hemi = pd.read_csv('./Data/Male/genes_AD_X.csv', sep=';')
```

```python
selected_AR = selected_AR[['complete','Feature Importance']]
selected_AD = selected_AD[['complete','Feature Importance']]
selected_hemi = selected_hemi[['complete','Feature Importance']]
selected_AR.rename(columns={'complete':'feature'},inplace=True)
selected_AD.rename(columns={'complete':'feature'},inplace=True)
selected_hemi.rename(columns={'complete':'feature'},inplace=True)
selected_AR.dropna(inplace=True)
selected_AD.dropna(inplace=True)
selected_hemi.dropna(inplace=True)
```

```python
selected_m = selected_hemi.feature.tolist() + selected_AD.feature.tolist() + selected_AR.feature.tolist() + ['grading']
```

```python
df_males_rare = df[df['Gender (M=0, F=1)']==0][selected_m]
```

```python
df_males_rare
```

```python
df_males_rare.to_csv('./Data/AssociationRules/Rare_Male_bool.csv', index=False)
```

```python
df_males_ar = df[df['Gender (M=0, F=1)']==0][selected_AR.feature.tolist() + ['grading']]
df_males_ar.to_csv('./Data/AssociationRules/AR_Male_bool.csv', index=False)
```

```python
df_males_ad = df[df['Gender (M=0, F=1)']==0][selected_AD.feature.tolist() + ['grading']]
df_males_ad.to_csv('./Data/AssociationRules/AD_Male_bool.csv', index=False)
```

```python
df_males_hemi = df[df['Gender (M=0, F=1)']==0][selected_hemi.feature.tolist() + ['grading']]
df_males_hemi.to_csv('./Data/AssociationRules/Hemi_Male_bool.csv', index=False)
```
