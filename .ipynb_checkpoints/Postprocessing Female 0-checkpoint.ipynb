```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import scipy as sp
import matplotlib.pyplot as plt
import re
from os import listdir
from os.path import isfile, join
from scipy import stats
from matplotlib.ticker import MaxNLocator

import warnings
warnings.filterwarnings('ignore')
```

```python
important = ['EP300','HERC5','TP53','HAPLN3','ALCAM','APOBEC1']
```

```python
df = pd.read_csv("./Results/rules_females_0.csv")
```

```python
#Split the string by reading the left hand and the right hand sides 
df['lhs'] = df['rules'].apply(lambda x : re.search(r"\{(.+)\} ", x).group(1))
df['rhs'] = df['rules'].apply(lambda x : re.search(r" \{(.+)\}", x).group(1))

#drop the rules column
df.drop(columns='rules',inplace = True)
#sort the columns differently 
df = df[['lhs','rhs','support','confidence','lift','count','coverage','pvalue_test']]

df.reset_index(inplace=True)
df.drop(columns='index', inplace=True)
df['lhs'] = df['lhs'].str.replace('=1','')
df['count'] = df['count'].astype('int32')
df = df.sort_values(by=['support'], ascending=False)
```

```python
minconf_1 = 0.8
minsup_1 = 0.12

df[(df['confidence']>=minconf_1) 
                   & (df['support']>=minsup_1)
                   & (df['rhs']=='grading=1')
                  ]
```

```python
minconf_0 = 0.8
minsup_0 = 0.12

df[(df['confidence']>=minconf_0) 
                   & (df['support']>=minsup_0)
                   & (df['rhs']=='grading=0')
                  ]
```

# Importo geni di suscettibilitÃ  e protezione

```python
feature_importance = pd.read_csv('./Data/Female/relevant genes in females.csv', sep=';')
```

```python
feature_importance['complete'] = feature_importance['complete'].astype('str')
feature_importance[feature_importance['Type']=='Common Homozygous']['complete'] = feature_importance[feature_importance['Type']=='Common Homozygous']['complete'].apply(lambda x : "{}{}".format(x,'_homo'))
feature_importance['Feature Importance'] = feature_importance['Feature Importance'].str.replace(',','.')
feature_importance['Feature Importance'] = pd.to_numeric(feature_importance['Feature Importance'],errors='coerce')
```

```python
feature_importance.loc[feature_importance['Type']=='Common Homozygous']['complete'] = feature_importance[feature_importance['Type']=='Common Homozygous']['complete'].apply(lambda x :  x +'_homo')
```

```python
feature_importance.complete = feature_importance.apply(lambda row: row['complete'] + '_homo' if row['Type'] == 'Common Homozygous' else row['complete'], axis=1)
```

```python
sus = feature_importance[feature_importance['Feature Importance']>=0]['complete'].tolist()
#sus = sus +['IRX5_1']
```

```python
prot = feature_importance[feature_importance['Feature Importance']<0]['complete']
```

```python
series = []
genes1 = []
genes0 = []
patients = []

for i in range(len(grading0)):
    split = grading0.iloc[i]['lhs'].split(',')
    g0 = []
    g1 = []
    for gene in split:
        if '=0' in gene:
            g0.append(gene.split('=')[0])
        elif '=1' in gene:
            g1.append(gene.split('=')[0]) 
    if (len(g0)>0 and len(g1)>0):
        boolean_values = (pat0[g1].values == 1) & (pat0[g0].values == 0)
    elif (len(g0)>0):
        boolean_values = (pat0[g0].values == 0)
    elif (len(g1)>0):
        boolean_values = (pat0[g1].values == 1)
    #print(len(pat0))
    #print(len(boolean_values))
    #print(boolean_values)
    series.append(pd.Series(boolean_values.all(1), name='bools'))
    patients.append(pat0[series[i].values].index.values)
```

```python
distinct = patients[0]

for i in range(1,len(patients)):
    distinct = np.union1d(distinct,patients[i])
    
print("Le regole piÃ¹ significative coprono: ",len(distinct), " pazienti non severi distinti su ",len(pat0))
print("Corrispondente al {perc:.2f}% dei pazienti gravi".format(perc=(len(distinct)/len(pat0))*100))
print("Corrispondente al {perc:.2f}% del totale".format(perc=(len(distinct)/len(pat))*100))
```

# Patients in common

```python
series = []
genes = []
patients = []
common_pat_0 = np.full((len(sel_prot),len(sel_prot)),0)
out = []

for i in range(len(sel_prot)):
    for j in range (len(sel_prot)):
        rule1 = grading0.iloc[i]['lhs'].split(',')
        rule2 = grading0.iloc[j]['lhs'].split(',')
        rule = list(set(rule1).union(set(rule2)))
        s = pd.Series((pat0[rule].values == 1).all(axis=1), name='bools')
        common_pat_0[i,j] = len(pat0[s.values])
        out.append([grading0.iloc[i]['lhs'],grading0.iloc[j]['lhs'],len(pat0[s.values])])
        #patients.append(pat0[series[i].values].index.values)
```
